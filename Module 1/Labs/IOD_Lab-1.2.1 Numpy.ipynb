# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# get matplot to show charts inline
%matplotlib inline
````

````python
np.random.seed(0)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

````python
print(height.size)
````

````python
np.min(height)
````

**8. Create histograms of height and weight**

````python
num_bins = 30
plt.hist(height, num_bins, facecolor='red', alpha=1)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height Distribution')
plt.show()
````

````python
plt.hist(weight, num_bins, facecolor='red', alpha=1)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
````

**9. Create separate boxplots of height and weight**

````python
plt.boxplot(height, showmeans = True, showfliers = True)
plt.xlabel('Distribution')
plt.ylabel('Height')
plt.title('Boxplot Distribution of Height')
plt.show()
````

````python
plt.boxplot(weight, showmeans = True, showfliers = True)
plt.xlabel('Distribution')
plt.ylabel('Weight')
plt.title('Boxplot Distribution of Weight')
plt.show()
````

**10. Create a scatterplot of height and weight**

````python
plt.scatter(height, weight, s = 5, facecolors = 'none', edgecolors = 'b')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Population Scatter Distribution: Weight vs Height')
plt.show()
````
