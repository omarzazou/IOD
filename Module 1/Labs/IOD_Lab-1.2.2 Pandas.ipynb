
````python
def row_count(df, col):
    number_of_rows = df[col].value_counts()
    return number_of_rows
````

# 5. Visualisation


#### 5.A Plot a histogram of `age`

````python
num_bins = 47
plt.hist(df['age'], num_bins)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
````

#### 5.B BoxPlot of `age`

````python
plt.boxplot(df.age, showmeans = True, showfliers = True)
plt.xlabel('Distribution')
plt.ylabel('Age')
plt.title('Boxplot Distribution of Age')
plt.show()
````

#### 5.C Create a scatterplot of `age` and `length_of_service`

````python
plt.scatter(df.age, df.length_of_service, s = 30, facecolors = 'none', edgecolors = 'b')
plt.xlabel('Age')
plt.ylabel('Length of Service')
plt.title('Length of Service vs Age')
plt.show()
````

#### 5.D Print a statistical summary of `age`
