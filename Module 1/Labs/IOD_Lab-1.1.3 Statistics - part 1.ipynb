# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# frequency of some observations using pie chart
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency_aus = [15, 32, 12, 7]
frequency_usa = [4, 10, 16, 28]
fig1, ax1 = plt.subplots(1, 2)
explode = (0, 0, 0.1, 0)  # exercise: modify this to "explode" only the 3rd slice (i.e. 'Data Scientists')

ax1[0].pie(frequency_aus, explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=False, startangle=90)
ax1[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1[1].pie(frequency_usa, explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=False, startangle=90)
ax1[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

fig1.tight_layout(pad=0)
#fig1.set_title('IT Roles Distribution in Australia & USA')
plt.show()
```

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))                               #this creates a list from 0 to 5
performance = [10,8,6,4,2,1]

# Create a bar plot of performance for each object
plt.bar(y_pos, performance)
plt.xticks(y_pos, objects)

# Use ylabel and title to label the y axis and provide a title
plt.ylabel('Students in favour of')
plt.title('Student Progrtamming Language Preferences')
plt.show()
```
